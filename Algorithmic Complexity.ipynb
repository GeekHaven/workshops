```python
def words(text): 
    return re.findall('[a-z]+', text.lower()) 

def train(features):
    model = collections.defaultdict(int)
    for f in features:
        model[f] += 1
    return model

NWORDS = train(words(file('big.txt').read()))
```

```python
sorted_words = sorted(NWORDS.items(), key=operator.itemgetter(0))
```

```python
def howmany(word, word_list):
    if word_list:
            midpoint =  len(word_list) / 2
            midword = word_list[midpoint][0]
            if len(word_list) == 1:
                return word_list[midpoint][1]
        if word > midword:
            return howmany(word, word_list[midpoint:])
        if word < midword:
            return howmany(word, word_list[:midpoint]) 
        if word == midword:
            return word_list[midpoint][1]
    else:
        return 'not in our wordlist'
    
def timehowmany(word, wordlist):
        start_time = datetime.now()
        result = howmany(word,wordlist)
        end_time = datetime.now()
        print result, ' in ', start_time-end_time
```
